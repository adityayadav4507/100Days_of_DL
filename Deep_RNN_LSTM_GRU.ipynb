```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
# load datasets
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)

# padding
x_train = pad_sequences(x_train, maxlen=100)
x_test =  pad_sequences(x_test, maxlen=100)
```

```python
# define RNN
model=Sequential([
    Embedding(10000, 32), # convert word into vector
    SimpleRNN(5, return_sequences=True), # RNN layer of 5 units
    SimpleRNN(2),                       # another RNN layer of 5 unit
    Dense(1, activation='sigmoid')      # output layer of classifiaction
])
model.build(input_shape=(None, 100))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# train model
history=model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
from tensorflow.keras.layers import GRU , LSTM
```

```python
# define LSTM
model=Sequential([
    Embedding(10000, 32), # convert word into vector
    LSTM(5, return_sequences=True), # LSTM layer of 5 units
    LSTM(2),                       # another LSTM layer of 5 unit
    Dense(1, activation='sigmoid')      # output layer of classifiaction
])
model.build(input_shape=(None, 100))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# train model
history=model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

```python
# define GRU
model=Sequential([
    Embedding(10000, 32), # convert word into vector
    GRU(5, return_sequences=True), # GRU layer of 5 units
    GRU(2),                       # another GRU layer of 5 unit
    Dense(1, activation='sigmoid')      # output layer of classifiaction
])
model.build(input_shape=(None, 100))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# train model
history=model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```
